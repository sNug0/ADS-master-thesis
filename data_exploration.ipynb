```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import librosa

path = ""
root = os.path.join(os.getcwd(), path)

# Create dataframe with filename, length and size
df = pd.DataFrame(columns=['filename', 'length', 'size'])

# Walk through all files in path and append filename, length and size to dataframe
for subdir, dirs, files in os.walk(root):
    for file in files:
        filepath = subdir + os.sep + file
        if filepath.endswith(".wav"):
            df = df.append({'filename': file, 'length': librosa.get_duration(filename=filepath), 'size': os.path.getsize(filepath)}, ignore_index=True)

# Plot histogram of length
plt.figure(figsize=(10, 5))
plt.hist(df['length'], bins=20)
plt.xlabel('Length (s)')
plt.ylabel('Count')
plt.title('Histogram of length')
plt.show()

# Plot histogram of size
plt.figure(figsize=(10, 5))
plt.hist(df['size'], bins=20)
plt.xlabel('Size (bytes)')
plt.ylabel('Count')
plt.title('Histogram of size')
plt.show()

# Plot boxplot of length
plt.figure(figsize=(10, 5))
plt.boxplot(df['length'])
plt.ylabel('Length (s)')
plt.title('Boxplot of length')
plt.show()

# Plot boxplot of size
plt.figure(figsize=(10, 5))
plt.boxplot(df['size'])
plt.ylabel('Size (bytes)')
plt.title('Boxplot of size')
plt.show()

# Plot scatterplot of length vs size
plt.figure(figsize=(10, 5))
plt.scatter(df['length'], df['size'])
plt.xlabel('Length (s)')
plt.ylabel('Size (bytes)')
plt.title('Scatterplot of length vs size')
plt.show()

# Calculate total length of all files
total_length = df['length'].sum()
print('Total length of all files: ' + str(total_length) + ' seconds')

# Calculate total length of all files in minutes
total_length_min = total_length / 60
print('Total length of all files: ' + str(total_length_min) + ' minutes')

# Calculate total size of all files
total_size = df['size'].sum()
print('Total size of all files: ' + str(total_size) + ' bytes')

# Calculate average length of all files
avg_length = df['length'].mean()
print('Average length of all files: ' + str(avg_length) + ' seconds')

# Calculate average length of all files in minutes
avg_length_min = avg_length / 60
print('Average length of all files: ' + str(avg_length_min) + ' minutes')

# Calculate average size of all files
avg_size = df['size'].mean()
print('Average size of all files: ' + str(avg_size) + ' bytes')

# Save dataframe to csv
df.to_csv('data_exploration.csv', index=False)


```
